# EDA for Cryotheraphy dataset

Importing required python libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

Reading data from specific columns of excel file

```python
cryo = pd.read_excel('C:\Users\Admin\Downloads\Cryotherapy.xlsx', usecols=['age','Time'], index = False )
```

Shape of dataset

```python
cryo.shape
```

Columns selected from Cryotheraphy excel file

```python
cryo.columns
```

Sample data of dataset

```python
cryo.sample(10)
```

Bar graphs

```python
#plt.show(plt.hist(cryo['age'], color='b'))
#plt.show(plt.hist(cryo['Time'], color='g'))

bg = cryo[['age','Time']].plot(kind='bar', title ="Cryogenic bar graph", figsize=(22,15), legend=True )
bg.set_xlabel("age", fontsize=22)
bg.set_ylabel("Time", fontsize=20)
plt.show()
```

Scatter plots using pandas

```python
cryo.plot.scatter(x='age', y='Time', c='age',s=100, edgecolor='None')
plt.show()
```

scatter plots using Matplotlib

```python
plt.scatter(cryo['age'],cryo['Time'], s=100)

plt.show()
```

```python
plt.plot(cryo['age'],cryo['Time'], color='g')
plt.show()
```
